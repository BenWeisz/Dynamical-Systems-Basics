```python
# Please install these modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 2 # The end of our range
N = 100 # The number of points we will use
t = np.linspace(0, T, N)
```

### Solution to system with $c = 1$

```python
c = 1
xt_c1 = c * np.exp(t)
```

```python
plt.plot(t, xt_c1, label="x(t)")

# Add axis labels
plt.xlabel("time (s)")
plt.ylabel("x(t)")

plt.grid(True)
plt.legend()

plt.title("x(t) = ce^x with c = 1")

plt.savefig("method1-fig1.png")
plt.show()
```

### Solution to system with $c = x(0)$ for $x(0) = 3$

```python
c = 3
xt_c3 = c * np.exp(t)
```

```python
plt.plot(t, xt_c3, label="x(t)")

# Add axis labels
plt.xlabel("time (s)")
plt.ylabel("x(t)")

plt.grid(True)
plt.legend()

plt.title("x(t) = x(0)e^x with x(0) = 3")
ax = plt.gca()
ax.set_yticks(np.arange(0, 23, 2) + 1)

plt.savefig("method1-fig2.png")
plt.show()
```
